```python
#-*- coding: utf-8 -*-
#

import pandas as pd

#
inputfile = '../data/sales_data.xls'
data = pd.read_excel(inputfile, index_col = u'') #

#
#10
data[data == u''] = 1
data[data == u''] = 1
data[data == u''] = 1
data[data != 1] = 0
x = data.iloc[:,:3].as_matrix().astype(int)
y = data.iloc[:,3].as_matrix().astype(int)

from keras.models import Sequential
#from keras.layers.core import Dense, Activation
from keras.layers import Dense, Activation

model = Sequential() #
model.add(Dense(input_dim = 3, output_dim = 10))  # :model.add(Dense(3,10))
model.add(Activation('relu')) #relu
model.add(Dense(input_dim = 10, output_dim = 1))  #model.add(Dense(10,1))
model.add(Activation('sigmoid'))  #0-1sigmoid

model.compile(loss = 'binary_crossentropy', optimizer = 'adam')#
#binary_crossentropy
#binarymean_squared_errorcategorical_crossentropy()
#adamsgdrmsprop

#nb_epochbatch_size
model.fit(x, y, nb_epoch = 1000, batch_size = 10) 
yp = model.predict_classes(x).reshape(len(y)) #
```
